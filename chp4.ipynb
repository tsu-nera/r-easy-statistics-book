```r
標本平均 <- numeric(length=10000) #推定値を格納する場所を予約
for(i in 1:10000){ #中括弧に囲まれた処理を 10000回繰り返す
    標本 <- rnorm(n=100,mean=50,sd=10) #N(50,10^2)からn=100の標本を抽出
    標本平均[i] <- mean(標本) #標本平均を計算する
}
```

```r
var(標本平均)
```

```r
hist(標本平均)
```

```r
標本分散 <- numeric(length=10000) #推定値を格納する場所を予約
不偏分散 <- numeric(length=10000) #推定値を格納する場所を予約
for(i in 1:10000){ #中括弧に囲まれた処理を 10000回繰り返す
    標本 <- rnorm(n=10,mean=50,sd=10) #N(50,10^2)からn=10の標本を抽出
    標本分散[i] <- mean((標本-mean(標本))^2) #標本分散を計算する
    不偏分散[i] <- var(標本) #不偏分散を計算する
}
```

```r
mean(標本分散)
```

```r
mean(不偏分散)
```

```r
sd(標本分散)
```

```r
sd(不偏分散)
```

```r
hist(標本分散,breaks=seq(0,500,10))
hist(不偏分散,breaks=seq(0,500,10))
```

```r
標本分散誤差100以上 <- ifelse(標本分散>=200,1,0)
不偏分散誤差100以上 <- ifelse(不偏分散>=200,1,0)
```

```r
table(標本分散誤差100以上)
```

```r
table(不偏分散誤差100以上)
```

```r
mean(sqrt(不偏分散))
```

```r
標本平均 <- numeric(length=10000) #推定値を格納する場所を予約
標本中央値 <- numeric(length=10000) #推定値を格納する場所を予約
for(i in 1:10000){ #中括弧に囲まれた処理を 10000回繰り返す
    標本 <- rnorm(n=10,mean=50,sd=10) #N(50,10^2)から n=10の標本を抽出
    標本平均[i] <- mean(標本) #標本平均を計算する
    標本中央値[i] <- median(標本) #標本中央値を計算する
}
```

```r
mean(標本平均)
```

```r
mean(標本中央値)
```

```r
sd(標本平均)
```

```r
sd(標本中央値)
```
