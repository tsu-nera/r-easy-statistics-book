```r
mean(x); median(x)
```

```r
mean(y); median(y)
```

```r
varp <- function(x) {
    標本分散 <- var(x)*(length(x)-1)/length(x)
    標本分散
}
```

```r
varp(x); varp(y)
```

```r
sqrt(varp(x)); sqrt(varp(y))
```
